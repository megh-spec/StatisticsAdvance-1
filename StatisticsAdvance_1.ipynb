1. What is a random variable in probability theory?
   - A random variable (RV) in probability is a function that assigns a numerical value to each outcome of a random experiment, linking theoretical probability to real-world data, like assigning '1' to 'Heads' and '0' to 'Tails' in a coin flip. RVs quantify uncertain results, are usually denoted by capital letters (X, Y), and can be discrete (countable values, e.g., dice roll) or continuous (values in an interval, e.g., height).
2. What are the types of random variables ?
   - Random Variable - GeeksforGeeksThere are two primary types of random variables: Discrete, which take countable, distinct values (like coin flips or dice rolls), and Continuous, which can take any value within an interval (like height or temperature). A less common third type, the Mixed Random Variable, combines both discrete and continuous elements.
3. Explain the difference between discrete and continuous distributions.
   - Discrete distributions handle countable, separate outcomes (like dice rolls, coin flips, counts) with non-zero probabilities for each value, using bars on graphs; continuous distributions handle any value within a range (like height, temperature, time) with zero probability for a single point, using smooth curves, with probability found by area under the curve.
4.  What is a binomial distribution, and how is it used in probability?
    - A binomial distribution models the probability of getting a specific number of "successes" in a fixed number of independent trials, where each trial has only two outcomes (success/failure) with a constant probability of success.
5. What is the standard normal distribution, and why is it important?
   - The standard normal distribution is a continuous probability distribution with a mean equal to 0 and a standard deviation equal to 1. It is a special case of the normal distribution and is denoted by Z ~ N(0, 1). The curve of the standard normal distribution is bell-shaped and symmetric about the mean.

It is important because it allows any normal random variable to be converted into a standard form using the Z-score. This makes probability calculations easy by using a single standard normal table. It is widely used in statistics for hypothesis testing, confidence intervals, comparison of data, and many other statistical analyses.

6. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
   - The Central Limit Theorem (CLT) states that when a sufficiently large number of independent random samples are taken from any population with a finite mean and variance, the distribution of the sample mean approaches a normal distribution, regardless of the shape of the original population distribution.

It is critical in statistics because it allows statisticians to use normal distribution methods even when the population distribution is not normal. CLT forms the basis of many statistical procedures such as hypothesis testing, confidence intervals, and estimation. It makes it possible to make reliable inferences about a population using sample data.

7. What is the significance of confidence intervals in statistical analysis?
   -Confidence intervals are used in statistical analysis to estimate a range of values within which the true population parameter is likely to lie. Instead of giving a single estimate, a confidence interval provides an upper and lower bound along with a confidence level, such as 95%.

They are significant because they measure the reliability and precision of an estimate, show the uncertainty associated with sample data, and help in making informed decisions. Confidence intervals are widely used in hypothesis testing, research studies, and data analysis to draw conclusions about a population based on sample results.

8. What is the concept of expected value in a probability distribution?
   - The expected value is the average or mean value of a random variable in a probability distribution. It represents the long-run outcome that is expected if an experiment is repeated many times. The expected value is calculated by multiplying each possible value of the random variable by its probability and then adding all the results.
  
   The expected value is important because it summarizes the distribution with a single representative value and is widely used in statistics, economics, decision-making, and risk analysis.

● Explain how you would apply the Central Limit Theorem to estimate the average sales

● Write the Python code to compute the mean sales and its confidence interval.

    - Even if daily sales data is not perfectly normally distributed, the Central Limit Theorem states that for a sufficiently large sample size, the sampling distribution of the sample mean will be approximately normal.

Here, we treat the available sales data as a sample. Using CLT, we can:

1. Compute the sample mean as an estimate of average daily sales.

2. Compute the standard error of the mean.

3. Use the normal distribution (Z = 1.96) to construct a 95% confidence interval for the true average daily sales.

```python
#10. You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend. daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270,265, 255, 250, 260]
import numpy as np
import math

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

n = len(daily_sales)
mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)

z = 1.96
margin_error = z * (std_sales / math.sqrt(n))

ci_lower = mean_sales - margin_error
ci_upper = mean_sales + margin_error

print("Mean Daily Sales:", mean_sales)
print("95% Confidence Interval:", (ci_lower, ci_upper))

```
